```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
Data = pd.read_excel("C:/Users/nxt-1/Desktop/Dataset/laptopData_after_cleaning.xlsx")
```

```python
Data.set_index("Labtop_ID")
```

```python
Data.isnull().sum()
```

```python
Data.dropna(inplace=True)
```

```python
Data["Company"].unique()
```

```python
Label = preprocessing.LabelEncoder()
```

```python
Lable_Company = Label.fit_transform(Data["Company"])
Lable_TypeName = Label.fit_transform(Data["TypeName"])
Lable_OpSys = Label.fit_transform(Data["OpSys"])
```

```python
Data ["Lable_Company"] = Lable_Company
Data["Lable_OpSys"] = Lable_OpSys
Data["Lable_TypeName"] = Lable_TypeName
```

```python
Data.head(20)
```

```python
Data.isnull().sum()
```

```python
i3 = np.array(Data["Cpu"][Data["Cpu"].str.contains("i3")])
i5 = np.array(Data["Cpu"][Data["Cpu"].str.contains("i5")])
i7 = np.array(Data["Cpu"][Data["Cpu"].str.contains("i7")])
```

```python
Data["Cpu"].replace(i3,"i3",inplace=True)
Data["Cpu"].replace(i5,"i5",inplace=True)
Data["Cpu"].replace(i7,"i7",inplace=True)
```

```python
Data["Cpu"].replace(i9,"i9",inplace=True)
```

```python
Data["Cpu"].value_counts().sort_values(ascending=False)
```

```python
Lable_Cpu_type = Label.fit_transform(Data["Cpu"])
Data["Lable_Cpu_type"] = Lable_Cpu_type 
```

```python
Features = ["Lable_Company","Lable_TypeName","Lable_Cpu_type","Lable_OpSys","Weight","Ram","Inches"]
x = Data[Features]
y = Data["Price"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,train_size=0.25)
```

```python
Model_LR = LinearRegression(positive=True)
Model_DT = DecisionTreeRegressor()
Model_SVR = SVR()
```

```python
Model_LR.fit(X_train,Y_train)
```

```python
Model_DT.fit(X_train,Y_train)
```

```python
Model_SVR.fit(X_train,Y_train)
```

```python
print("The last score before add cpu type of Linear Regression model are : ",Model_LR.score(X_test,Y_test)*100)
print("The last score before add cpu type of Decision Tree Regressor model are : ",Model_DT.score(X_test,Y_test)*100)
print("The last score before add cpu type of SVR are :",Model_SVR.score(X_test,Y_test)*100)
```

```python
print("The last score after add cpu type of Linear Regression model are : ",Model_LR.score(X_test,Y_test)*100)
print("The last score after add cpu type of Decision Tree Regressor model are : ",Model_DT.score(X_test,Y_test)*100)
print("The last score after add cpu type of SVR are :",Model_SVR.score(X_test,Y_test)*100)
```
